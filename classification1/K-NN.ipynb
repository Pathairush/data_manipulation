## K-Nearest Neighbor

n_neighbors is a hyperparameter to tune

```python
import numpy as np
from sklearn import neighbors, datasets
from sklearn import preprocessing 
 
n_neighbors = 3
 
# import some data to play with
iris = datasets.load_iris()
 
# prepare data
X = iris.data[:, :2]  
y = iris.target
 
# we create an instance of Neighbours Classifier and fit the data.
clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
clf.fit(X, y)
 
# make prediction
sl = input('Enter sepal length (cm): ')
sw = input('Enter sepal width (cm): ')
dataClass = clf.predict([[sl,sw]])
print('Prediction: '),
 
if dataClass == 0:
    print('Iris Setosa')
elif dataClass == 1:
    print('Iris Versicolour')
else:
    print('Iris Virginica')
```

```python
import matplotlib.pyplot as plt
plt.scatter(x = X[:,0],y = X[:,1],c = y);
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the iris dataset
x = iris.data[:,:]  

# Split the data into training/testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.60, random_state=42) # using 4 features
# x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.60, random_state=42) # using only 2 features

# adding list of k params
params_grid = np.array(range(1,11))
acc = [];

for n_neighbors in params_grid:
    # instantiate learning model (k = 3)
    knn = neighbors.KNeighborsClassifier(n_neighbors)

    # fitting the model
    knn.fit(x_train, y_train)

    # predict the response
    pred = knn.predict(x_test)

    # evaluate accuracy
    acc_test = accuracy_score(y_test, pred)
    acc.append(acc_test);

    print("n_neighbors = " + str(n_neighbors) +"\naccuracy score for test dataset: {} %".format('%0.2f'%(acc_test*100)))
```

```python
plt.scatter(y = acc, x = params_grid ,label = "% acc")
plt.legend(loc="best")
```
